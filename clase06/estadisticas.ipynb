#Tipos de variables

Las variables son características o atributos que pueden tomar diferentes valores. Es crucial comprender los tipos de variables para analizar datos y construir modelos adecuados.

# Clasificacion



## 1. Variables Cualitativas (Categóricas):

* Estas variables representan cualidades o atributos no numericos.

**Nominales:**
Las categorías no tienen un orden inherente.<br>

- color de ojos (azul, marrón, verde)
- estado civil (soltero, casado, divorciado)
- tipo de sangre (A, B, AB, O)
- Nacionalidad
- Aprobado / Desaprobado


**Ordinales:** Las categorías tienen un orden o jerarquía.

- nivel de educación (primaria, secundaria, terciaria)
- nivel de satisfacción (bajo, medio, alto)
- clasificaciones de una competencia (primero, segundo, tercero)



## 2. Variables Cuantitativas (Numéricas):

* Estas variables representan cantidades numéricas.

**Discretas:** Toman valores enteros y contables.

- número de hijos
- número de habitaciones en una casa
- número de llamadas telefónicas recibidas

**Continuas:** Pueden tomar cualquier valor dentro de un intervalo.

- altura
- peso
- temperatura
- tiempo

#Fundamentos de Estadisticas

## Estadística Descriptiva

La **estadística descriptiva** es una rama de la estadística que **se encarga de recopilar, organizar, describir y analizar un conjunto de datos** con el fin de extraer información relevante. Su objetivo principal es ofrecer una visión general y comprensible de los datos, lo que permite identificar patrones, tendencias y relaciones dentro del conjunto de datos.

La importancia de la estadística descriptiva radica en su capacidad para **proporcionar un resumen claro y conciso de los datos, lo cual es esencial para la toma de decisiones informadas**. Al utilizar medidas de resumen como la media, mediana, moda, varianza y el desvio estándar, podemos obtener una comprensión más profunda de la distribución de los datos y de las características principales del conjunto de datos analizado.

#### Medidas de tendencia central

**1. Media**: Promedio aritmetico de los datos

La media es una medida de tendencia central que se calcula sumando todos los valores de un conjunto de datos y dividiendo esta suma por la cantidad de valores. Es uno de los indicadores más utilizados para resumir un conjunto de datos, ya que ofrece una idea general del valor promedio.

```python
# mean
muestra.mean()
```

**Uso de la Media:** La media es útil cuando se necesita un valor representativo del conjunto de datos. Sin embargo, puede ser influenciada por valores atípicos (outliers), lo que puede distorsionar su interpretación.

**2. Mediana**: El valor central de los datos ordenados

La mediana es el valor que ocupa la posición central en un conjunto de datos ordenados de menor a mayor. Si el número de observaciones es impar, la mediana es el valor en la posición central. Si es par, la mediana es el promedio de los dos valores centrales.

```python
# median
muestra.median()
```

```python
# cantidad de elementos impares
pd.Series([1, 2, 3, 4, 5]).median()
```

```python
# cantidad de elementos pares
pd.Series([1, 2, 3, 4, 5, 100]).median()
```

```python
# como influye en la media los valores atipicos
pd.Series([1, 2, 3, 4, 5, 6]).median()
```

```python
pd.Series([1, 2, 3, 4, 5, 6]).mean()
```

```python
pd.Series([1, 2, 3, 4, 5, 60]).mean()
```

```python
pd.Series([1, 2, 3, 4, 5, 60]).median()
```

**Uso de la Mediana:** La mediana es especialmente útil en distribuciones asimétricas o cuando existen valores atípicos, ya que no se ve afectada por estos como la media.

**3. Moda**: El valor que parece con mayor frecuencia

La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Un conjunto de datos puede tener una única moda (unimodal), más de una moda (multimodal) o ninguna moda si todos los valores son únicos.

```python
muestra
```

```python
# mode()
muestra.mode()
```

**Uso de la Moda:** La moda se utiliza principalmente en variables cualitativas o en situaciones donde se requiere conocer el valor más común dentro del conjunto de datos.

**2. Varianza:** Una medida de cuánto se dispersan los datos alrededor de la media

La varianza mide la dispersión de los datos respecto a la media. Se calcula promediando el cuadrado de las diferencias entre cada valor y la media. La varianza ofrece una idea de cuán dispersos están los datos; una varianza alta indica que los datos están muy dispersos, mientras que una varianza baja indica que los datos están más concentrados alrededor de la media.

```python
# var
muestra.var()
```

```python
# ejemplos
pd.Series([1, 2, 3, 4, 5, 4, 2, 2]).var()
```

```python
pd.Series([1, 2, 3, 4, 5, 7, 2, 2]).mean()
```

**Uso de la Varianza:** La varianza es fundamental para entender la variabilidad dentro de un conjunto de datos, y es una base para el cálculo de otras medidas como la desviación estándar.

**3. Desviación estándar:** La raíz cuadrada de la varianza, que indica la dispersión promedio de los datos.

La desviación estándar es la raíz cuadrada de la varianza y proporciona una medida de dispersión en las mismas unidades que los datos originales. Es una medida más intuitiva que la varianza porque refleja directamente la magnitud de las desviaciones respecto a la media.

```python
# std
muestra.std()
```

**Uso de la Desviación Estándar:** La desviación estándar es ampliamente utilizada para entender la dispersión de los datos y es fundamental en muchas aplicaciones estadísticas, incluyendo la evaluación de la volatilidad en mercados financieros o la variabilidad en procesos de producción.

**4. Percentiles y Cuartiles:** Los valores que dividen los datos ordenados en cien partes iguales.

Los cuartiles y percentiles son medidas que dividen un conjunto de datos ordenados en partes iguales, proporcionando información detallada sobre la distribución de los datos. Los **cuartiles** dividen los datos en **cuatro partes iguales**: el primer cuartil (Q1) marca el 25% de los datos más bajos, el segundo cuartil (Q2), que corresponde a la mediana, divide el conjunto en la mitad, y el tercer cuartil (Q3) marca el 75% de los datos más bajos. Los **percentiles**, por su parte, dividen los datos en **100 partes iguales**, y son especialmente útiles para interpretar la posición relativa de un valor dentro de un conjunto de datos. Por ejemplo, si un valor se encuentra en el percentil 90, significa que el 90% de los valores en el conjunto son menores o iguales a él.

- Q1 = P25
- Q2 = P50 (la mediana)
- Q3 = P75

```python
# quantile
muestra.quantile([0.25, 0.5, 0.75])
```

```python
# describe
muestra.describe()
```

**Uso de Cuartiles y Percentiles**: Estas medidas son fundamentales para análisis en los que se requiere comprender la dispersión y distribución de los datos más allá de la media y la mediana, permitiendo identificar rangos y establecer comparaciones entre diferentes subconjuntos de datos.

# Distribución de variables

## Distribuciones uniformes

Una distribución uniforme es aquella en la que **todos** los valores posibles dentro de un rango tienen la **misma probabilidad** de ocurrir. En otras palabras, cada resultado es igualmente probable. Esta distribución es a menudo representada por una forma rectangular en un histograma, donde la altura de cada barra es igual, reflejando que todas las categorías tienen la misma frecuencia.

- lanzamimiento de un dado
- numero en la ruleta

## Distribuciones normales

La distribución normal, también conocida como la campana de Gauss, es una de las distribuciones más importantes en estadística. Se caracteriza por ser simétrica respecto a su media y tener una forma de campana. La mayor parte de los datos se agrupa alrededor de la media, y la probabilidad disminuye a medida que nos alejamos de la media en ambos sentidos. La media, mediana y moda coinciden en el centro de la distribución.

- peso recien nacidos
- altura de la pobleción

## Diferencias entre Distribuciones Uniformes y Normales

La principal diferencia entre una distribución uniforme y una normal es la forma de la distribución de los datos. Mientras que la distribución uniforme muestra que todos los resultados son igualmente probables, la distribución normal indica que los resultados cercanos a la media son más probables, y los extremos son menos probables.

- ***Forma:*** La distribución uniforme es rectangular, mientras que la distribución normal tiene forma de campana.

- ***Simetría:*** Aunque ambas distribuciones pueden ser simétricas, la distribución normal tiene una concentración de datos alrededor de la media, mientras que la uniforme no tiene tal concentración.

## Interpretación de Histogramas

Un histograma es una representación gráfica de la distribución de un conjunto de datos. En un histograma, el eje horizontal (X) muestra las categorías o intervalos de los datos, y el eje vertical (Y) muestra la frecuencia o número de observaciones en cada categoría.

- ***Interpretación de Histogramas en Distribuciones Uniformes:*** En un histograma de una distribución uniforme, se espera ver barras de igual altura, indicando que cada intervalo tiene aproximadamente el mismo número de observaciones.

- ***Interpretación de Histogramas en Distribuciones Normales:*** En un histograma de una distribución normal, se espera ver una forma de campana, con las barras centrales más altas, que representan la mayor concentración de datos alrededor de la media, y barras más bajas a medida que nos alejamos de la media.

```python
# Generar datos de una distribución normal

import scipy.stats as stats

mu = 0
sigma = 1
datos = np.random.normal(mu, sigma, 1000)

# Crear una Serie de pandas
serie = pd.Series(datos)

# Calcular la media y la desviación estándar
media = serie.mean()
desviacion_estandar = serie.std()

# Visualizar la distribución con un histograma
serie.hist(bins=30, density=True, alpha=0.6, color='b')

# Curva de distribución normal teórica
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), 'r')

plt.title('Distribución Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
```

# Interpretación de la Correlación Lineal y el Coeficiente de Correlación

La correlación lineal se refiere a la relación de linealidad entre dos variables. Esta relación se mide a través del coeficiente de correlación, que toma valores entre -1 y 1:

- Un coeficiente de correlación cercano a 1 indica una fuerte correlación positiva, lo que significa que, en general, a mayores valores de una variable le corresponden mayores valores de la otra.

- Un coeficiente cercano a -1 indica una fuerte correlación negativa, lo que significa que a mayores valores de una variable le corresponden menores valores de la otra.

- Un coeficiente cercano a 0 sugiere que no hay una correlación lineal fuerte entre las dos variables, es decir, los cambios en una variable no están asociados de manera consistente con cambios en la otra.

Es crucial recordar que la correlación no implica causalidad. Aunque dos variables puedan mostrar una relación aparente en un diagrama de dispersión, esto no significa necesariamente que una cause la variación en la otra. Para hacer afirmaciones causales, se requiere un análisis más profundo, como la regresión y otros métodos estadísticos más avanzados.

Los diagramas de dispersión son una herramienta poderosa para descubrir relaciones y patrones ocultos en los datos, pero siempre deben interpretarse con cuidado y en el contexto adecuado.

```python
# Datos de ejemplo (altura en cm, peso en kg)
datos = {'Altura': [160, 165, 170, 175, 180, 185],
         'Peso': [60, 65, 70, 75, 80, 85]}
df = pd.DataFrame(datos)

# Calcular la correlación
correlacion = df['Altura'].corr(df['Peso'])
print(f"Correlación entre altura y peso: {correlacion:.2f}")

# Diagrama de dispersión
sns.scatterplot(x='Altura', y='Peso', data=df)
plt.title('Correlación entre altura y peso')
plt.show()
```
