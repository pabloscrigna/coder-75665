Modo API

```python
# metodo plt
plt.figure(figsize=(12, 6))

plt.title('Mi primer grafico')

plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
plt.figure(figsize=(12, 6))
plt.title('Mi segundo grafico')

plt.subplot(1, 2, 1)  # fila, columna, panel seleccionado
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Linea vertical'])
plt.xlabel('X')
plt.ylabel('X al cuadrado')

plt.subplot(1, 2, 2)     # fila columna panel
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Linea Horizontal'])
plt.xlabel('X')
plt.ylabel('-X Squared')


plt.xlabel('X')
plt.ylabel('X al cuadrado')
```

Modo OOP (API orientada a objetos)

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3,
    marker='o', markersize=8, label='X^2')

axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X al 2')

axes.set_title("Mi primer OOP grafico")

axes.legend()

fig
```

```python
df = pd.DataFrame({"Year": [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010],
                   "Population1": [2.51, 2.65, 2.79, 2.92, 3.00, 3.08, 3.09, 3.12, 3.20, 3.25]})
```

Grafico de barras

```python
paises = ['Argentina', 'Bolivia', 'Brasil', 'Chile', 'Colombia', 'Ecuador', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela', "Canada"]
np.random.seed(10)
poblacion = [np.random.randint(10,100)+i for i in range(len(paises))]
```

```python
paises
```

```python
poblacion
```

```python
df = pd.DataFrame({"pais": paises, "poblacion": poblacion})
```

```python
df.head()
```

```python
# df.bar grafico de barras
plt.figure(figsize=(10,6))
plt.title("Poblacion mundial")
plt.xlabel("Paises")
plt.xticks(rotation=90)
plt.ylabel("Poblacion")
plt.grid()
plt.bar(df["pais"], df['poblacion'])

for indice, valor in enumerate(df["poblacion"]):
    plt.text(indice, valor+1, str(valor), ha="center")
plt.show()
```

```python
df = df.sort_values(by="poblacion", ascending=False)
```

```python
# Nota agregar distintos colores en cada barra
# df.bar grafico de barras
plt.figure(figsize=(10,6))
plt.title("Poblacion mundial")
plt.xlabel("Paises")
plt.xticks(rotation=90)
plt.ylabel("Poblacion")
plt.grid()
plt.bar(df["pais"], df['poblacion'], color="green")

for indice, valor in enumerate(df["poblacion"]):
    plt.text(indice, valor+1, str(valor), ha="center")
plt.show()
```

Grafico de Tortas

```python
# Datos de ejemplo -- Unidades de vehiculos vendidas por marca
data = {
    "Fabricante": ["Toyota", "Volkswagen", "Ford", "Honda", "BMW", "Chevrolet"],
    "Unidades": [25000, 20000, 15000, 13000, 12000, 15000]
}

# Crear DataFrame
df = pd.DataFrame(data)
df
```

```python
# grafico de torta o pie
plt.figure(figsize=(10,6))
plt.pie(df["Unidades"], labels=df["Fabricante"], autopct="%1.1f%%")
plt.title("Unidades vendidas por fabricante")
plt.show()
```

```python
# grafico de torta o pie
plt.figure(figsize=(10,6))
plt.pie(
    df["Unidades"],
    labels=df["Fabricante"],
    autopct="%1.1f%%",
    startangle=90,
    textprops={"fontsize": 12}
)
plt.title("Unidades vendidas por fabricante")
plt.show()
```

```python
# Vamos a ordenarlo
df = df.sort_values(by="Unidades", ascending=False)
```

```python
df
```

```python
# grafico de torta o pie
plt.figure(figsize=(10,6))
plt.pie(
    df["Unidades"],
    labels=df["Fabricante"],
    autopct="%1.1f%%",
    startangle=90,
    textprops={"fontsize": 12},
    colors = ["red", "green", "blue", "yellow", "orange", "purple"]
)
plt.title("Unidades vendidas por fabricante")
plt.show()
```

Histogramas

```python
# tiempo de respuesta de un servidor web
tiempos_respuesta = np.random.normal(loc=200, scale=50, size=1000)  # Media=200ms, Desv
```

```python
# histograma plt.hist
plt.figure(figsize=(10,6))
plt.hist(tiempos_respuesta, bins=50, alpha=0.7)
plt.show()
```

```python
# histograma plt.hist
plt.figure(figsize=(10,6))
plt.hist(tiempos_respuesta, bins=100, alpha=0.7, density=True)
plt.show()
```

Graficos de Dispersion o de puntos (scatterplot)

```python
np.random.seed(42)

# Simular datos de horas de estudio y calificación en examen
horas_estudio = np.random.randint(1, 11, 100)  # Horas de estudio (1 a 10 horas)
calificaciones = horas_estudio * 10 + np.random.randint(0, 5, 100)  # Relación lineal + ruido
```

```python
df = pd.DataFrame({"horas_estudio": horas_estudio, "calificaciones": calificaciones})
```

```python
# grafico de dispersion : plt.scatter
plt.figure(figsize=(10,6))
plt.scatter(df["horas_estudio"], df["calificaciones"], alpha=0.5)
```
